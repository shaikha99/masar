```python
#import the needed packages
import seaborn as sn
import pandas as pd

#read the data
df = pd.read_csv('https://raw.githubusercontent.com/DSomair/Masar_DS/main/Datasets/Week2/diabetes_unclean.csv')
df
```

```python
# get the column
df.columns
```

```python
#rename columns
df.rename(columns={'No_Pation':'Pat_ID'}, inplace=True)
```

```python
#check how many missing values
df.isnull().sum()
```

```python
#replace the missing value(s) with mean

df['HbA1c'] = df['HbA1c'].fillna(df['HbA1c'].mean())
```

```python
df.info()
```

```python
#check the class column values
df['CLASS'].unique()
```

```python
#replace the white spaces
df.CLASS = df.CLASS.str.replace('Y ', 'Y')
df.CLASS = df.CLASS.str.replace('N ', 'N')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Set up the matplotlib figure
plt.figure(figsize=(10, 6))

# Create boxplots for each column using seaborn
sns.boxplot(data=df
            , orient='v', palette='Set2')

# Set plot labels and title
plt.xlabel('Columns')
plt.ylabel('Values')
plt.title('Boxplot for Each Column')

# Show the plot
plt.show()
```
